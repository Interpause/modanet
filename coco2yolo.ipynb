```python
from pylabel import importer
```

```python
coco_val = importer.ImportCoco(
    './dataset/modanet_instances_val.json',
    name='modanet_val'
)
#https://github.com/pylabel-project/pylabel/issues/35
coco_val.df = coco_val.df[coco_val.df.cat_id != 'nan']
coco_val.df['split'] = 'val'

coco_train = importer.ImportCoco(
    './dataset/modanet_instances_train.json',
    name='modanet_train'
)
coco_train.df = coco_train.df[coco_train.df.cat_id != 'nan']
coco_train.df['split'] = 'train'
```

```python
coco_val.export.ExportToYoloV5(output_path='yolo/labels', use_splits=True, copy_images=True)
coco_train.export.ExportToYoloV5(output_path='yolo/labels', use_splits=True, copy_images=True)
```
